```r
library('xml2')

# xml_viewを使うため
library('htmltidy')

# xmlToDataFrameを使うため
library('XML')
```

```r
# ニコニコ動画の動画情報を取得できるAPI。返り値はXML
url = 'http://ext.nicovideo.jp/api/getthumbinfo/sm37512100'
```

```r
data = read_xml(url)
```

```r
# 中身を表示する
xml_view(data)
```

```r
# 構造だけ表示する
xml_structure(data)
```

```r
# データフレームとして読み込む
df = xmlToDataFrame(url)
```

```r
#df
# 構造は失われる
# 重複する要素は結合される
```

```r
# 構造を見る
# 子要素を列挙

data %>% xml_children() 
```

```r
# 子ノードの子ノード

data %>% xml_children() %>% xml_children() 
```

```r
# xml_path()を使うことでXpath記法で表示
data %>% xml_children() %>% xml_children()  %>% xml_path()
```

```r
# xpath記法を使って要素を取得
data %>% xml_find_all('/nicovideo_thumb_response/thumb/video_id') %>% xml_text()
```

```r
# すべての子ノードから"tag"をすべて抜き出す
xml_find_all(data, '//tag')
```

```r
# 列挙した中から9番目の要素を取得
xml_find_all(data, '//tag[position()=9]')
```

```r
# タグの中から属性lockの値を抜き出す
xml_find_all(data, '//tag') %>% xml_attr('lock')
```

```r
# Yahoo! Japanのトップページ
url = 'https://www.yahoo.co.jp/'
```

```r
html = read_html(url)
```

```r
# rootノードを確認
html %>% xml_path()
```

```r
# 見出し(h1タグ)を取得
html %>% xml_find_all('//h1') %>% xml_text()
```

```r
# リンク(aタグ)をすべて抜き出す
html %>% xml_find_all('//a')
```

```r
# タグからリンクを抽出
html %>% xml_find_all('//a') %>% xml_attr('href')
```

```r
# こう書いてもよい (@で属性を指定)
html %>% xml_find_all('//a/@href') %>% xml_text()
```

```r
# リンクが張られている文字列
html %>% xml_find_all('//a') %>% xml_text()
```

```r
# idがtoollistである要素を入手する
html %>% xml_find_first('id("ToolList")') %>% xml_children() 
```

```r
html %>% xml_find_first('id("ToolList")') %>% xml_children() %>% xml_children() %>% xml_text()
```
